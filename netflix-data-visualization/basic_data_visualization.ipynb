# Question 2)
## If a producer wants to release some content, which month must he do so?

```python
# Cleaning and preparing for visualization of heatmap
```

```python
netflix_overall.isnull().sum()
```

```python
netflix_date = netflix_overall[["date_added"]].dropna()
```

```python
netflix_date.isnull().values.any()
```

```python
# Seperating months and years in the date_added column
```

```python
netflix_date["year"] = netflix_date["date_added"].apply(lambda x : x.split(', ')[-1])
```

```python
netflix_date["month"] = netflix_date["date_added"].apply(lambda x : x.lstrip().split(' ')[0])
```

```python
netflix_date
```

```python
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"][::-1]
month_order
```

```python
df = netflix_date.groupby("year")["month"].value_counts().unstack().fillna(0)[month_order].T
df
```

```python
# Creating custom color map
```

```python
colors = [[0.13, 0.12, 0.12], [0.90, 0.05, 0.08], [0.96, 0.96, 0.95]]
cmap_name = 'MyColormap'
cm = LinearSegmentedColormap.from_list(cmap_name, colors, N=256)
```

```python
# Visualization
```

```python
plt.figure(figsize = (8, 5), dpi=130)
plt.pcolor(df,cmap = cm, edgecolors = 'white', linewidths = 1)
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns, fontsize = 8)
plt.yticks(np.arange(0.5, len(df.index), 1), df.index, fontsize = 8)

plt.title('Netflix Contents Update', fontsize = 13, fontweight = 'bold', position = (0.2, 0.2))
cbar = plt.colorbar()

cbar.ax.tick_params(labelsize = 8)
cbar.ax.minorticks_on()
```

## Months that less produced contents can be a good choice for produce something. According to 2020 and 2021, February and March were the months comparatively much less content was released. Therefore, these months may be a good choice for produce contents.

# Question 4)
## Which are the top 10 countries with the highest movie content production?

```python
netflix_overall.dropna()
```

```python
# Preparing for visualization
```

```python
# Seperating countries of each row
```

```python
countries = {}
netflix_movies["country"] = netflix_movies["country"].fillna("Unknown")
cou = list(netflix_movies["country"])
for i in cou:
    i = list(i.split(','))
    if len(i) == 1:
        if i in list(countries.keys()):
            countries[i] +=1
        else: countries[i[0]] = 1
    else:
        for j in i:
            if j in list(countries.keys()):
                countries[j]+= 1
            else:
                countries[j] = 1
```

```python
countries_fin = {}
for country,no in countries.items():
    country = country.replace(' ','')
    if country in list(countries_fin.keys()):
        countries_fin[country] += no
    else:
        countries_fin[country] = no
        
countries_fin = {k: v for k, v in sorted(countries_fin.items(), key = lambda item: item[1], reverse = True)}
```

```python
# Visualization
```

```python
plt.figure(figsize = (6,6))
plt.title("Countries producing movie content")
ax = sns.barplot(x = list(countries_fin.keys())[0:10], y = list(countries_fin.values())[0:10], palette=['#221f1f', '#3b1d1d', '#551b1b', '#701a1a', '#8b1818', '#a61616', '#c11414',
                                                                                                        '#dc1212', '#f5100f', '#ff0f0f', '#ff1d1a', '#ff2a26', '#ff3632', '#ff433f'])
ax.set_xticklabels(list(countries_fin.keys())[0:10], rotation = 90);
```
